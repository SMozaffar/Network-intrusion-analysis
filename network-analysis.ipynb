# Step 2: Model Fitting

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc, precision_recall_curve

# Splitting the data into training and testing sets
X = df.drop(columns=['label'])
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Dictionary to hold models
models = {
    'GaussianNB': GaussianNB(),
    'RandomForest': RandomForestClassifier(random_state=42),
    'MLP': MLPClassifier(random_state=42),
    'SVM': SVC(probability=True, random_state=42),
    'DecisionTree': DecisionTreeClassifier(random_state=42)
}

# Dictionary to store results
model_results = {}

def evaluate_model_multiclass(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred, output_dict=True)
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    return accuracy, report, conf_matrix

# Evaluating models with multiclass support
for name, model in models.items():
    accuracy, report, conf_matrix = evaluate_model_multiclass(model, X_train, X_test, y_train, y_test)
    model_results[name] = {
        'accuracy': accuracy,
        'report': report,
        'conf_matrix': conf_matrix
    }

# Displaying the model accuracies
{model_name: model_results[model_name]['accuracy'] for model_name in model_results}
```

Let's take a closer look at the Random Forest and Decision Tree models trained on the data.

```python
from sklearn.tree import plot_tree
from sklearn.preprocessing import LabelEncoder 

# Encode the target labels
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['label'])

class_names_str = [str(label) for label in label_encoder.classes_]

# Visualize a part of the Decision Tree
plt.figure(figsize=(25, 10))
plot_tree(models["DecisionTree"], filled=True, max_depth=3, feature_names=X.columns, class_names=class_names_str)
plt.title("Decision Tree Visualization (Depth = 3)")
plt.show()
```

The decision tree visualization provides a window into the model's decision-making process.

- **Root Node**:

    - The root node splits based on the flag feature, which indicates that the model first considers the status of the network connection as the most critical factor in its initial classification.

- **Subsequent Splits**:

    - The tree then splits on features such as src_bytes, same_srv_rate, and dst_host_srv_count. These splits show the model's approach to differentiating between normal and attack traffic, refining its decisions based on the volume of traffic, connection rates, and other related factors.

    - For instance, low src_bytes might indicate less data being sent, possibly indicative of specific types of attacks or legitimate traffic.

- **Class Assignments**:

    - Leaf nodes are assigned to classes like nmap, mailbomb, and portsweep, demonstrating how the model identifies specific types of attacks based on the paths it takes through the tree.

```python
import numpy as np

# Extract feature importances
importances = models["RandomForest"].feature_importances_
indices = np.argsort(importances)[::-1]

# Plot the feature importances
plt.figure(figsize=(12, 8))
plt.title("Feature Importances from Random Forest")
plt.bar(range(10), importances[indices[:10]], align="center")
plt.xticks(range(10), [X.columns[i] for i in indices[:10]], rotation=90)
plt.xlim([-1, 10])
plt.show()
```

The feature importance of the Random Forest model provides insights into which features are most influential in the classification process.

- **Top Features**:
    - *src_bytes*: The most important feature, likely indicating that the number of bytes sent from the source is a strong indicator of whether the traffic is normal or an attack.
    - *dst_bytes*: Similarly, the number of bytes sent to the destination is important, possibly distinguishing between different types of attacks or normal traffic.
    - *flag*: This feature, which indicates the status of the connection (e.g., normal or error), is also highly important, as certain flags might be more common in attack scenarios.

Other important features include dst_host_srv_count, dst_host_same_srv_rate, service, same_srv_rate, duration, count, and dst_host_rerror_rate. These features likely capture various aspects of network traffic behavior, such as connection rates and error rates, which may help in distinguishing between normal and malicious activities.
